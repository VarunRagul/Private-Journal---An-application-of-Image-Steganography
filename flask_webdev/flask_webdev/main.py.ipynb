```python
from flask import Flask, render_template, request
from flask_sqlalchemy import SQLAlchemy



app = Flask(__name__)
app.config['SQLALCHEMY_DATABASE_URI'] = 'mysql://root:@localhost/journal_en'
app.config['SQLALCHEMY_TRACK_MODIFICATIONS'] = True
db = SQLAlchemy(app)


class Encrypt(db.Model):
    Sno = db.Column(db.Integer, primary_key=True)
    Name = db.Column(db.String(25), nullable=False)
    Journal_Entry = db.Column(db.String(2000000), nullable=False)
    Image = db.Column(db.LargeBinary(), nullable=False)


class Decrypt(db.Model):
    Sno = db.Column(db.Integer, primary_key=True)
    Name = db.Column(db.String(25), nullable=False)
    Image = db.Column(db.LargeBinary(), nullable=False)

    
@app.route("/")
def home():
    return render_template('index.html')


@app.route("/encrypt", methods=['GET','POST'])
def encrypt():
    print("entered enrypt")
    if(request.method=='POST'):
        #add entry to db
        
        print("encrypt post")
        name=request.form.get('name')
        journal=request.form.get('message')
        image=request.form.get('image')
        
        entry=Encrypt(Name=name, Journal_Entry=journal, Image=image)
        db.session.add(entry)
        db.session.commit()
    return render_template('encrypt.html')



@app.route("/decrypt", methods=['GET','POST'])
def decrypt():
    print("entered decrypt")
    if(request.method=='POST'):
        #add entry to db
        
        print("decrypt post")
        name=request.form.get('name')
        image=request.form.get('image')
        
        entry=Decrypt(Name=name, Image=image)
        db.session.add(entry)
        db.session.commit()
                      
    return render_template('decrypt.html')
```

```python
#Encode data into image
def encoding(img, journal):
    
   
        
    from PIL import Image
    from IPython.display import display
    
    image=Image.open(img, 'r')
    print("The image to be encoded:")
    display(image)
    
    def get_num_pixels(filepath):
        width,height=Image.open(filepath).size
        return width*height

    y=get_num_pixels(image)
    
    data=input("Type in your journal entry : ")
    x=len(data)
    if(x==0):
        raise ValueError('Data is empty')
    if(y>(x*3)):
        print("The image has ",y," number of pixels, You're good to go")
    else:
        print("Error, Number of pixels is not enough, upload a larger image.")
        exit()
 
    #making a copy of the input image
    newimg=image.copy()
    
    #modify all the neccessary pixels with the data
    encode_enc(newimg, data)
    
    #save the encoded image under the name that the user specifies
    new_img_name = input("Enter the name of new image(with extension) : ")
    newimg.save(new_img_name, str(new_img_name.split(".")[1].upper()))
    print("Encoding has finished successfully, you can now download your new image!")
```

```python
#Convert encoding data into 8-bit binary using ASCII value
def generateData(data):
        newdata=[]
#ord gives ascii value
#format(ord(i),'08b' adds leading zeros to the binary of ito make it 8 bits

        for i in data:
            newdata.append(format(ord(i), '08b'))
        return newdata
```

```python
def modifyPix(pix,data):
 
    #datalist ['01101000', '01101001', '01101001'] looks like this
    datalist=generateData(data)
    lendata=len(datalist)
    #iter creates an iterator for the pixels to iterate one element at a time
    imdata=iter(pix)
 
    for i in range(lendata):
    #meaning, if lendata = 3, i iterates from 0 to 2 i.e., 0,1,2
    #Extracting 3 pixels at a time
        pix=[value for value in imdata.__next__()[:3] + imdata.__next__()[:3] + imdata.__next__()[:3]]
 
        # Pixel value should be made
        # odd for 1 and even for 0
        for j in range(0, 8):
            if (datalist[i][j] == '0' and pix[j]% 2 != 0):
                pix[j] -= 1
 
            elif (datalist[i][j] == '1' and pix[j] % 2 == 0):
                if(pix[j] != 0):
                    pix[j] -= 1
                else:
                    pix[j] += 1
    
 
        # Eighth pixel of every set tells whether to stop or read further
        # 0 means keep reading, 1 means message is over
        
        #if reached the last character of the input text, then make the last ele of the last pixel odd
        if (i==lendata-1):
            if (pix[-1] % 2 == 0):
                if(pix[-1] != 0):
                    pix[-1] -= 1
                else:
                    pix[-1] += 1
 
        #if we have not reached the end yet, make the last ele of the pixel even(0 means keep reading)
        else:
            if (pix[-1] % 2 != 0):
                pix[-1] -= 1
 
        #storing multiple values in a single variable -> tuple    
        pix = tuple(pix)
    
        #yield is used to return from a function without destroying states of its local variable when the function is called 
        #the execution starts from the last yield statement.
        yield pix[0:3]
        yield pix[3:6]
        yield pix[6:9]
```

```python
def encode_enc(newimg,data):
    #finding size of the image given as input
    w=newimg.size[0]
   
    #initializing coordinates to exchange existing pixels with new ones
    (x,y)=(0,0)
    # for example something like 0x0000026E11CD52D0
    for pixel in modifyPix(newimg.getdata(),data):
        
        #Putting modified pixels in the new image
        newimg.putpixel((x,y),pixel)
        
        #if pixels on the first column/row are done, we move to the next column/row, make x=0, increment y 
        if (x==w-1):
            x=0
            y+=1
        #if we haven't reached the end of the first column/row of pixels, keep incrementing x
        else:
            x+=1
```

```python
#Decode the data in the image
def decoding(img):
    
    from PIL import Image
    from IPython.display import display
   
    image = Image.open(img, 'r')
    print("The image to be decoded:")
    display(image)
    data = ''
    
    #iterating through the pixels of the image
    imgdata=iter(image.getdata())
 
    #separating pixels in groups of three
    while(True):
        pixels=[value for value in imgdata.__next__()[:3] + imgdata.__next__()[:3] + imgdata.__next__()[:3]]
 
        # string of binary data
        binstr=''
        #reading the binary vaue from the pixels, based on odd and even numbers
        for i in pixels[:8]:
            if (i%2==0):
                binstr+='0'
            else:
                binstr+='1'
         
        #appending the digits, and converting to int followed by character
        #stop and return the data if the last pixel value is odd
        data+=chr(int(binstr,2))
        if (pixels[-1] % 2 != 0):
            return data
```
